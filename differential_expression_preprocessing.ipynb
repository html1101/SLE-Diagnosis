```python
# Imports
import pandas as pd # Used to analyze the contents of the datasets.
import os # Used to find the files in ../data.
import seaborn as sns # Used for data visualization.
from scipy import stats
sns.set_theme(style="darkgrid")
```

**(1)** Load Data from ../data.

```python
# a) Find files in ../data.
path = "../data"
files = os.listdir(path)

# Make sure we have at least 1 file to analyze.
assert len(files) >= 1, "Error: Not enough files to analyze!"

# b) Read contents of each file.
contents = pd.read_excel(F"{path}/{files[0]}")
contents = contents.set_index("Unnamed: 0")
for i, filename in enumerate(files[1:]):
    print(F"Reading contents of file {path}/{filename}")
    content = pd.read_excel(F"{path}/{filename}")

    # Create a large concatenated file containing all our files. Might remove genes like A1CF that not all datasets have expression data for.
    contents = contents.join(content.set_index("Unnamed: 0"), on="Unnamed: 0", how="left").dropna(axis=0, how="any")

contents
```

```python
# Convert to large excel file so we don't need to load each one of them every time we want to load this data.
contents.to_excel("differential_expression_remove.xlsx")
```

```python
# Run zscore on each column
contents = pd.read_excel("differential_expression_remove.xlsx").fillna(0).set_index("Gene Name")
contents
```

```python
# Now save it with z scores across gene axis
attempt = pd.DataFrame(data=stats.zscore(contents, axis=1), columns=contents.columns, index=contents.index)
attempt
```

```python
attempt.to_excel("differential_expression_zscores.xlsx")
```
